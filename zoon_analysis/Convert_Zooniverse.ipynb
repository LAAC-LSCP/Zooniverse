```python
import sys, os
import numpy as np
import pandas as pd
import json
```

```python
project_name = "maturity-of-baby-sounds"

classification_file = project_name + "-classifications-last.csv"

print(classification_file)
```

```python
classifications_all = pd.read_csv(classification_file,delimiter=",",encoding='utf-8')
n_class = len(classifications_all)

print("File %s read with %d rows." % (classification_file, n_class))
```

```python
classifications_all.head()
```

```python
annot=classifications_all['annotations']
subj_data=classifications_all["subject_data"]
usr_id=classifications_all["user_id"]
ids=classifications_all["subject_ids"]


i=0
data=[]
for v_,f_,id_ in zip(annot,subj_data,usr_id):
    #load string in json format
    jdata_v = json.loads(v_)
    jdata_f = json.loads(f_)
   # print(jdata_v,jdata_f)
    #read as dictionary
    values_dict=jdata_v[0]
    files_dict=jdata_f[str(ids[i])]
    #create dataset and convert to pandas df
    try:
        data.append([id_,files_dict["name"],values_dict["value"]])
    #Get around UTF-8-sig encoding issue
    except:
        try:
            data.append([id_,files_dict['\ufeffname'],values_dict["value"]])
        except:
            try:
                data.append([id_,files_dict['Name'],values_dict["value"]])
            except:
                pass
    i+=1
    
    
df = pd.DataFrame(data, columns = ['UserID', 'AudioData','Answer'])
df['Dataset'] = 'Maturity of Baby Sounds'
df['Question'] = 'Please Classify this sound'
df
```

```python
def make_js(column):
    data_list=[]
    for s in column:
        js=json.loads(s)
        data_list.append(js)
    return data_list
        
js=make_js(classifications_all['metadata'])
for i in js[0:10]:
    print("started at: {}".format(i["started_at"]))
    print("finished at: {} \n".format(i["finished_at"]))
```

```python
users_all = classifications_all['user_name'].unique()
n_users = len(users_all)

# if the classification is from a classifier who isn't signed in, the user_name field has "not-logged-in-[user_ip]"
is_unreg = np.array([q.startswith("not-logged-in") for q in users_all])
is_reg   = np.invert(is_unreg)

n_unreg = sum(is_unreg)
n_reg   = sum(is_reg)

print("%d classifications from %d classifiers, of which %d (%.0f percent) were signed-in and %d (%.0f percent) were not signed in.\n" % (n_class, n_users, n_reg, (float(n_reg)/float(n_users)*100.), n_unreg, (float(n_unreg)/float(n_users)*100.)))

print("Average classifications per user: %.1f" % (float(n_class)/float(n_users)))
```
